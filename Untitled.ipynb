```python
#Load the neccesary libraries and load the dataset

#import the library needed
import pandas as pd

#load the dataset
df = pd.read_csv('toys.csv')
df.head()
```

```python
#convert categorical data to integer representations of unique categories (integer encoding)

#import the library needed
from sklearn.preprocessing import LabelEncoder

#using LabelEncoder, categorical labels will be transformed into numerical values
le = LabelEncoder()
categorical_data = ['Color', 'size', 'firmness', 'age']
for i in categorical_data:
    df[i] = le.fit_transform(df[i])

df.head()
```

```python
#split the data into features (X) and labels (y)

#create DataFrame X containing independent variables
X = df.drop(columns=['expensive'])
#Create DataFrame y containing dependent variables
y = df['expensive']

#split the data into training and testing sets.Specify 70% of the data to be used for training 
#import the library needed

from sklearn.model_selection import train_test_split

#split the dataset into 70% for training and 30% for testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)
```

```python
#train a decision tree classifier model
#import the library needed
from sklearn.tree import DecisionTreeClassifier

#Train the model
clf = DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
```

```python
#evaluates the performance of a trained decision tree classifier model using test data
# print the accuracy_score
# print the confusion_matrix

#import the library needed for the accuracy
from sklearn.metrics import accuracy_score, confusion_matrix

#evaluates the performance of a trained decision tree classifier model using test data
predictions = clf.predict(X_test)
print(predictions)

accuracy = accuracy_score(y_test, predictions)

#Print accuracy
print(accuracy)

#Print Consufion Matrix
confusion_matrix(y_test, predictions)#
```

```python
# plot the decision tree

#import the libraries needed
from sklearn import tree
from matplotlib import pyplot as plt

#Plot the decision tree
fig = plt.figure(figsize = (50,30))
_= tree.plot_tree(clf, feature_names = X.columns, class_names=clf.classes_, filled=True)
plt.show()
```
